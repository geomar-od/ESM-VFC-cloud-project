# Dask IO workload

Specify Dask cluster,

```python
import dask.distributed
```

```python
cluster = dask.distributed.LocalCluster(
    n_workers=1,
    threads_per_worker=2,
    memory_limit='10GB',
    ip='0.0.0.0'
)
```

```python
client = dask.distributed.Client(cluster); client
```

Connect pi workload zarr store,

```python
import dask.array
```

```python
zarr_store_base_path = 'gs://<bucket>'
xy_zarr_store = dask.array.from_zarr(zarr_store_base_path+'/pi-workload.zarr')
```

```python
xy_zarr_store
```

Calculate pi,

```python
def calculate_pi(xy_zarr_store):
    """Calculate pi using a Monte Carlo method."""
    xy_inside_circle = (xy_zarr_store ** 2).sum(axis=1) < 1
    pi = 4 * xy_inside_circle.mean()
    return pi
```

```python
pi = calculate_pi(xy_zarr_store)
```

```python
%%time
pi.compute()
```
